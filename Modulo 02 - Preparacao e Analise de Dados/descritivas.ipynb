```python
import numpy as np
import pandas as pd
```

```python
dados = [10, 12, 13, 15, 15, 15, 16, 17, 18, 20, 100]
# Convertendo para DataFrame para facilitar a visualização
df = pd.DataFrame(dados, columns=['valores'])
```

```python

# Calculando as estatísticas
media = np.mean(dados)
mediana = np.median(dados)
moda = df['valores'].mode()[0] 
desvio_padrao = np.std(dados)
```

```python
print("Dados originais:")
print(df)
print("\nEstatísticas:")
print(f"Média: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda}")
print(f"Desvio Padrão: {desvio_padrao:.2f}")
```

```python
print("\nExplicação:")
print("1. Média: É sensível a outliers (como o valor 100)")
print("2. Mediana: Não é afetada por outliers")
print("3. Moda: Mostra o valor mais frequente")
print("4. Desvio Padrão: Mede a dispersão dos dados")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(dados, 'ro', label='Dados')
plt.axhline(y=media, color='b', linestyle='-', label=f'Média: {media:.2f}')
plt.axhline(y=mediana, color='g', linestyle='-', label=f'Mediana: {mediana:.2f}')
plt.axhline(y=moda, color='y', linestyle='-', label=f'Moda: {moda}')
plt.title('Visualização das Medidas Estatísticas')
plt.legend()
plt.grid(True)
plt.show() 
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(dados)
plt.title('Boxplot dos Dados')
plt.ylabel('Valores')
plt.grid(True)
plt.show()
```
